```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = np.random.normal(10,2,1000)
plt.hist(x)
x = x.reshape(-1,1);
```

```python
y = np.random.normal(10,100,1000)
plt.hist(y)
y = y.reshape(-1,1);
```

```python
plt.xlim(0,max(x))
plt.ylim(0,max(y))
plt.scatter(x,y)
```

```python
model = LinearRegression(positive=True).fit(y,x)
m = model.coef_[0][0]
b = model.intercept_
f'y={m}x+{b}'
```

```python
x = x.reshape(1000)
y = y.reshape(1000)
plt.xlim(0,max(x))
plt.ylim(0,max(y))
plt.plot(x,y, 'o')
plt.plot(x, m*x+b)
```
